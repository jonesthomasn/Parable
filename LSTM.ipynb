```python
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
#from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```python
testdata = pd.read_csv("E:/parable_testdata.csv")
traindata = pd.read_csv("E:/parable_traindata.csv")
```

```python
testdata = testdata.dropna()
traindata = traindata.dropna()
```

```python
x_train = traindata[['Price','DebtPS','Debt2Eq','CurrentR','FCFPS','PE','PS','PB','EV2EBITDA','PM','EBITDA','Cash','MinorityInt','MkCap','Leverage','RevGrowth','ROIC','CFO','DilEPS','R3MEQI','ICBIndNo','ICBSecNo','CUSIP','NDebt2CF','NI2Profit','AcRec1yGr','InvDays','TDebt1yGr','d200Pxchange','BV5yGr','CEOTenure',]]
x_test = testdata[['Price','DebtPS','Debt2Eq','CurrentR','FCFPS','PE','PS','PB','EV2EBITDA','PM','EBITDA','Cash','MinorityInt','MkCap','Leverage','RevGrowth','ROIC','CFO','DilEPS','R3MEQI','ICBIndNo','ICBSecNo','CUSIP','NDebt2CF','NI2Profit','AcRec1yGr','InvDays','TDebt1yGr','d200Pxchange','BV5yGr','CEOTenure',]]
y_train = traindata['Bought']
y_test = testdata['Bought']
```

```python
x_train_tp = np.array(x_train[:1000]).T
y_train_tp = np.array(y_train[:1000]).T
x_train_tp = x_train_tp.reshape((1,31,1000))
y_train_tp = y_train_tp.reshape((1,1000))
x_train_tp.shape
```

```python
x_test_tp = np.array(x_train[1000:2000]).T
y_test_tp = np.array(y_train[1000:2000]).T
x_test_tp = x_test_tp.reshape((1,31,1000))
y_test_tp = y_test_tp.reshape((1,1000))
```

```python
model = Sequential()
model.add(LSTM(1000,input_shape=x_test_tp.shape[1:],return_sequences=False))
```

```python
#model.add(Dense(1000))
```

```python
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train_tp,y_train_tp,nb_epoch=1000,batch_size=1,verbose=2,validation_data=(x_test_tp,y_test_tp))
```

```python
prediction = model.predict(x_test_tp)
```
